```python
data = pd.read_csv('Monthly Ice Cream Shop Sales by Flavor.csv')
data
```

### line graph that displays total profts across all months
#### creating variable names for columns

```python
months = data['Month']
total_profits = data['Total Profit']
flavors = data.columns[1:7]
```

```python
plt.plot(months, total_profits, marker = 'o', markerfacecolor = 'red')
plt.xlabel('Month')
plt.ylabel('Profit $')
plt.title('Profit vs Month')
plt.show()
```

### Split the line graph by ice cream flavor to compare sales by flavors over time.

```python
plt.figure(figsize = (12,6))
for flavor in flavors:
    plt.plot(months, data[flavor], marker = 'o', label=flavor)

plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.title('Sales by Flavor vs Month')
plt.legend()
plt.show()
```

### Create a bar chart that displays the total number of units sold per month

```python
total_units = data['Total Units']
plt.bar(months, total_units)
plt.xlabel('Month')
plt.ylabel('Total Units Sold')
plt.title('Total Units Sold per Month')
plt.show()
```
